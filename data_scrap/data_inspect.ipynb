```python
import pandas as pd
import json

with open('../data/data.json', 'r') as file:
    data = json.load(file)
    df = pd.DataFrame(data)

print(df.isnull().values.any())
```

```python
import matplotlib.pyplot as plt

ratings_count = df.groupby(["rating_star"]).size().values
ratings = sorted(df["rating_star"].unique())

plt.bar(ratings, ratings_count)
plt.title('Ratings Count')
plt.xlabel('Ratings')
plt.ylabel('Count')
plt.show()
```
